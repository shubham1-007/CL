```python
import pandas as pd
df=pd.read_csv('realestate.csv')
df.head()
```

```python
df.columns=df.columns.str.strip()
```

```python
print(df.isnull().sum())
```

```python
df.dropna(inplace=True)
```

```python
filtered=df[(df['transactiondate']> 2013) & (df['stores']==5)] 
print(filtered)
```

```python
df_encoded=pd.get_dummies(filtered)
df_encoded.head()
```
