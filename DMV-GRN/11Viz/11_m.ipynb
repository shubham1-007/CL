```python
import pandas as pd
df=pd.read_csv('PNQ_AQI.csv')

df_shape=df.shape
df_col=df.columns
df_dtypes=df.dtypes
#print above vars
df.head()
```

```python
#3
relevant_features=['Date','SO2 µg/m3','Nox µg/m3','RSPM µg/m3','CO2 µg/m3','AQI']
df_rel=df[relevant_features]
df_rel
```

```python
#4
import matplotlib.pyplot as plt

# Create a line plot of the overall AQI over time
plt.plot(df_rel['AQI'],df_rel['Date'])

# Set the title and labels of the plot
plt.title("Overall AQI Trend over Time")
plt.xlabel("Date")
plt.ylabel("AQI")

# Add a legend to the plot
plt.legend()

# Show the plot
plt.show()
```

```python
# Plot the SO2 levels over time
plt.plot(df_rel['SO2 µg/m3'], label='SO2 µg/m3')

# Set the title and labels of the plot
plt.title("SO2 µg/m3 Trend over Time")
plt.xlabel("Date")
plt.ylabel("µg/m3")

# Add a legend to the plot
plt.legend()

# Show the plot
plt.show()

# Plot the NOx levels over time
plt.plot(df_rel['Nox µg/m3'], label='NOx µg/m3')

# Set the title and labels of the plot
plt.title("NOx µg/m3 Trend over Time")
plt.xlabel("Date")
plt.ylabel("µg/m3")

# Add a legend to the plot
plt.legend()

# Show the plot
plt.show()

# Plot the RSPM levels over time
plt.plot(df_rel['RSPM µg/m3'], label='RSPM µg/m3')

# Set the title and labels of the plot
plt.title("RSPM µg/m3 Trend over Time")
plt.xlabel("Date")
plt.ylabel("µg/m3")

# Add a legend to the plot
plt.legend()

# Show the plot
plt.show()

# Plot the CO2 levels over time
plt.plot(df_rel['CO2 µg/m3'], label='CO2 µg/m3')

# Set the title and labels of the plot
plt.title("CO2 µg/m3 Trend over Time")
plt.xlabel("Date")
plt.ylabel("µg/m3")

# Add a legend to the plot
plt.legend()

# Show the plot
plt.show()
```
