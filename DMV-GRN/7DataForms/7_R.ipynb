```python
import pandas as pd
csv_data=pd.read_csv('customers.csv')

excel_data=pd.read_excel('customers.xlsx',engine='openpyxl')

import json 
with open ('Customers.json','r') as json_file:
    json_data=json.load(json_file)
    json_data=pd.DataFrame(json_data)

print("This is csv data")
print(csv_data.head())

print("This is excel data")
print(excel_data.head())

print("This is json data")
print(json_data)
```

```python
csv_data.fillna(0,inplace=True)
csv_data.drop_duplicates(inplace=True)

excel_data.drop_duplicates(subset=['first_name', 'last_name'], inplace=Trueinplace=True)
```

```python
unified_data=pd.concat([csv_data,excel_data,json_data])
unified_data.head()
```
