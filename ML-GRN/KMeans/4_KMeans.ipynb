Implement K-Means clustering on Iris.csv dataset. Determine the number of clusters
using the elbow method.

Step 1: Importing the required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

Load the Iris dataset

```python
data = pd.read_csv("Iris.csv")
X = data.iloc[:, 1:5].values  # Selecting the features
print(X)
```

Standardize the features

```python
scaler = StandardScaler()
X_std = scaler.fit_transform(X)
X_std
```

Determine the optimal number of clusters using the elbow method

```python
wcss = []  # Within-cluster sum of squares
```

```python
for i in range(1, 15):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X_std)
    wcss.append(kmeans.inertia_)
```

Plot the elbow graph

```python
plt.plot(range(1, 15), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
print(wcss.shape)
```

Additional


```python
# Import the seaborn library
import seaborn as sns

# Create a KMeans object with 3 clusters
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)

# Fit the KMeans object to the standardized data
kmeans.fit(X_std)

# Predict the cluster labels for each data point
cluster_labels = kmeans.predict(X_std)

# Create a scatter plot of the data points, colored by cluster label
sns.scatterplot(x=X_std[:, 0], y=X_std[:, 1], hue=cluster_labels)

# Add a title and labels to the plot
plt.title('Final Clusters Formed Using K-Means Algorithm')
plt.xlabel('Sepal Length (Standardized)')
plt.ylabel('Sepal Width (Standardized)')

# Show the plot
plt.show()
```
